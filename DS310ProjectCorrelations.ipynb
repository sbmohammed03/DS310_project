```python
import pandas as pd

def calculate_correlations(df, target_df):
    #takes a df of covariates and calculates correlation to target col and returns a sorted df
    # Create an empty list to store dictionaries
    correlation_data = []

    # Iterate through each column in the DataFrame
    for column_name in df.columns:
        # Calculate the correlation between the current column and the target column
        corr = df[column_name].corr(target_df)

        # Add a dictionary containing the column name and corresponding correlation to the list
        correlation_data.append({'Policy': column_name, 'Correlation': corr})

    # Create a DataFrame from the list of dictionaries
    correlation_df = pd.DataFrame(correlation_data)

    # Sort by absolute value of correlation to identify strongest relationships
    correlation_df = correlation_df.sort_values(by='Correlation', key=abs, ascending=False)

    # Reset the index after sorting
    correlation_df = correlation_df.reset_index(drop=True)

    # Remove rows with NaN correlations
    correlation_df = correlation_df.dropna()

    return correlation_df

```

```python
cases_corr = calculate_correlations(policies, cases_growth)
cases_corr
```

```python
deaths_corr = calculate_correlations(policies, deaths_growth)
deaths_corr
```

```python
import pandas as pd

def sort_combined_correlation(cases_corr, deaths_corr):
    # Merge the two DataFrames on the 'Policy' column
    combined_df = pd.merge(cases_corr, deaths_corr, on='Policy', suffixes=('_cases', '_deaths'))
    # Combine the correlation values from both DataFrames into a new column
    combined_df['Combined_Correlation'] = combined_df.iloc[:,1] + combined_df.iloc[:,2]
    # Sort by absolute value of combined correlations
    combined_df = combined_df.sort_values(by='Combined_Correlation', key=abs, ascending=False)
    # Reset the index after sorting
    combined_df = combined_df.reset_index(drop=True)
    return combined_df

# Example usage:
combined_df = sort_combined_correlation(cases_corr, deaths_corr)
combined_df
```

```python

def sort_combined_rank(cases_corr, deaths_corr):
    # Add temporary 'Rank' columns based on the index values
    cases_corr['Rank_1'] = cases_corr.index
    deaths_corr['Rank_2'] = deaths_corr.index
    # Merge the two DataFrames on the 'Policy' column
    combined_df = pd.merge(cases_corr, deaths_corr, on='Policy', how='outer')
    # Calculate the combined rank by adding the ranks from both DataFrames
    combined_df['Combined_Rank'] = combined_df['Rank_1'] + combined_df['Rank_2']
    # Sort by the combined rank in ascending order
    combined_df = combined_df.sort_values(by='Combined_Rank')
    # Reset the index after sorting
    combined_df = combined_df.reset_index(drop=True)
    # Select only the 'Policy' and 'Combined_Rank' columns
    return combined_df[['Policy', 'Combined_Rank']]
combined_df_rank = sort_combined_rank(cases_corr, deaths_corr)
combined_df_rank
```

### Data Normalization
- Since some of our categorical variables have larger ranger than others, to prevent those from dominating the other features, we normalize them all to be of equal importance.

```python
#Min Max Scaling Such that Each is Between 0 and 1:
from sklearn.preprocessing import MinMaxScaler
columns_to_normalize = sw_policies.columns

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Fit the scaler to the data and transform it
sw_policies[columns_to_normalize] = scaler.fit_transform(sw_policies[columns_to_normalize])

#combine with growth rates
# Create a new DataFrame by concatenating policies with deaths_growth and cases_growth
sw_combined_df = pd.concat([sw_policies, deaths_growth, cases_growth], axis=1)
sw_combined_df
```
